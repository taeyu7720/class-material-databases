# Activity 3 - Aggregations

```python
import pandas as pd
import sqlite3
db = sqlite3.connect('database.sqlite')
```

**Best overall player**

```python
pd.read_sql("""
    SELECT P.id, P.player_name, MAX(A.overall_rating)
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
""", db)
```

**Player with most potential**

```python
pd.read_sql("""
    SELECT P.id, P.player_name, MAX(A.potential)
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
""", db)
```

**Strongest player**

```python
pd.read_sql("""
    SELECT P.id, P.player_name, MAX(A.strength)
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
""", db)
```

**Weakest Player**

```python
pd.read_sql("""
    SELECT P.id, P.player_name, MIN(A.strength)
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
""", db)
```

**Average rating of players**

```python
pd.read_sql("""
    SELECT AVG(A.overall_rating)
    FROM Player_Attributes A
""", db)
```

**Players in with AVG rating**

```python
pd.read_sql("""
    SELECT P.id, P.player_name
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
    WHERE A.overall_rating = 68
    LIMIT 20
""", db)
```

**Select DISTINCT will fix the duplicates problem**

```python
pd.read_sql("""
    SELECT DISTINCT P.id, P.player_name
    FROM Player P
    INNER JOIN Player_Attributes A ON (P.player_api_id = A.player_api_id)
    WHERE A.overall_rating = 68
    LIMIT 20
""", db)
```

**Match with most goals scored in total**

```python
pd.read_sql("""
SELECT M.id, m.date, L.name, T_Home.team_long_name, T_Away.team_long_name, MAX(M.home_team_goal + M.away_team_goal)
FROM Match M
INNER JOIN League L ON (M.league_id = L.id)
INNER JOIN Team T_Home ON (M.home_team_api_id = T_Home.team_api_id)
INNER JOIN Team T_Away ON (M.away_team_api_id = T_Away.team_api_id)
""", db)
```

Alternatively:

```python
pd.read_sql("""
SELECT M.id, m.date, L.name, T_Home.team_long_name, T_Away.team_long_name, M.home_team_goal + M.away_team_goal as GOAL_SUM, M.home_team_goal, M.away_team_goal

FROM Match M

INNER JOIN League L ON (M.league_id = L.id)
INNER JOIN Team T_Home ON (M.home_team_api_id = T_Home.team_api_id)
INNER JOIN Team T_Away ON (M.away_team_api_id = T_Away.team_api_id)

ORDER BY GOAL_SUM DESC

LIMIT 2
""", db)
```

```python
pd.read_sql("""
SELECT M.id, m.date, L.name, T_Home.team_long_name, T_Away.team_long_name, M.home_team_goal + M.away_team_goal as GOAL_SUM, M.home_team_goal, M.away_team_goal

FROM Match M

INNER JOIN League L ON (M.league_id = L.id)
INNER JOIN Team T_Home ON (M.home_team_api_id = T_Home.team_api_id)
INNER JOIN Team T_Away ON (M.away_team_api_id = T_Away.team_api_id)

ORDER BY GOAL_SUM DESC, M.home_team_goal DESC

LIMIT 2
""", db)
```

**Worst home loss**

```python
pd.read_sql("""
SELECT M.id, m.date, L.name, T_Home.team_long_name, T_Away.team_long_name, M.home_team_goal, M.away_team_goal

FROM Match M

INNER JOIN League L ON (M.league_id = L.id)
INNER JOIN Team T_Home ON (M.home_team_api_id = T_Home.team_api_id)
INNER JOIN Team T_Away ON (M.away_team_api_id = T_Away.team_api_id)

WHERE M.home_team_goal < M.away_team_goal

ORDER BY M.away_team_goal DESC

LIMIT 2
""", db)
```

**Worst Away loss**

```python
pd.read_sql("""
SELECT M.id, m.date, L.name, T_Home.team_long_name, T_Away.team_long_name, M.home_team_goal, M.away_team_goal

FROM Match M

INNER JOIN League L ON (M.league_id = L.id)
INNER JOIN Team T_Home ON (M.home_team_api_id = T_Home.team_api_id)
INNER JOIN Team T_Away ON (M.away_team_api_id = T_Away.team_api_id)

WHERE M.home_team_goal > M.away_team_goal

ORDER BY M.home_team_goal DESC

LIMIT 2
""", db)
```
